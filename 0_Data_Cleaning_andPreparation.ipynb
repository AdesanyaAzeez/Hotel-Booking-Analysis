### **Data Cleaning**
1. Get 'Booking Date'
    
    NB: Lead time the number of days between Booking and Arrival (i.e. 'Arrival Date' minus 'Booking Date')

    To get 'Booking Date', we need to bring our 'Arrival Date' together

```python
# Join the Arrival Date Columns, and convert to DateTime
htb['arrival_date'] = pd.to_datetime(
    htb['arrival_date_year'].astype(str) + '-' + 
    htb['arrival_date_month'].astype(str) + '-' + 
    htb['arrival_date_day_of_month'].astype(str), 
    format="%Y-%B-%d"
    )

# Drop the 'arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month', 'arrival_date_week_number'
htb = htb.drop(columns=['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month', 'arrival_date_week_number'])

# Preview data
htb.head()
```

```python
# New column: Subtract 'Lead Time' from 'Arrival Date' 
htb['booking_date'] = htb['arrival_date'] - pd.to_timedelta(htb['lead_time'], unit='D')

# Preview data
htb.head()
```

2. Replace Null values

From previous findings, 4 columns have null values
- For 'Agent': (The id of the agent who brought or refered user) Replace NULL values with 'No agent'
- For 'Company': Replace NULL values with 'Not provided'
- For 'Country': Replace NULL values with 'Not provided'
- For 'Children': Replace NULL values with zero (0)

```python
# Replace NULL values accordingly
htb['agent'] = htb['agent'].replace(np.nan, 'No agent')
htb[['company', 'country']] = htb[['company', 'country']].replace(np.nan, 'Not provided')
htb['children'] = htb['children'].replace(np.nan, 0).convert_dtypes()
htb
```

```python
# Confirm that there are no NULL values
htb.isna().sum()
```

5. Drop unneccessary column(s)

```python
# Delete columns that are not needed or useful for analysis
htb = htb.drop(columns=['s/n'])

'''
When deleting columns, it is important to note that columns that are be useful for identifying records
are to be kept even if they are not needed for analysis. 
This helps maintain the data quality and integrity.

Deleting them might lead to data inaccuracy when some operations like 'Deleting duplicates' are done.
'''
```

6. Rename columns for better readabilty

```python
# Rename column adr to average daily rate for better readability
htb = htb.rename(columns={
    'adr': 'avg_daily_rate'
})
```

7. Replace abbreviations in some columns with their full meanings for better understanding

```python
# Check for unique values for columns where values are to be replaced
print(f'Values in meal column: {htb['meal'].unique()}')
print(f'Values in distribution_channel column: {htb['distribution_channel'].unique()}')
```

```python
# Replace abbrevation in 'meal', 'distribution_channel' columns respectively
htb['meal'] = htb['meal'].replace({
    'BB': 'Bed & Breakfast',
    'FB': 'Full Board',
    'HB': 'Half Board',
    'SC': 'Self Catering'
})

htb['distribution_channel'] = htb['distribution_channel'].replace({
    'TA/TO': 'Travel Agent/Tour Operator',
    'GDS': 'Global Distribution System'
})
```

8. After data cleaning, cross-check for duplicates again

```python
# Recheck for duplicates
htb[htb.duplicated()]
```

```python
# Print no of duplicates before deleting them
print(f'Number of duplicates: {htb.duplicated().sum():,}')

# Remove duplicates from dataframe 'htb'
htb = htb.drop_duplicates()

# Print no of duplicates after deleting them
print(f'Number of duplicates after removal: {htb.duplicated().sum()}')
```
